```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib
```

```python
url = 'https://zenodo.org/record/7674076/files/Sengupta_Lazarus_2023_F2_a.csv?download=1'
dt = pd.read_csv(url)

dt.head()
```

```python
sns.set(font_scale=0.8)
custom = {'font': 'Arial',
               'xtick.direction': 'in',
               'ytick.direction': 'in',
               'xtick.top': 'True',
               'ytick.right': 'True'}

sns.set_style('ticks', rc = custom)
sns.set_style({'font.family': 'Arial'})
dt.set_index('Year').plot.line(legend=False, colormap='bone', figsize = (6, 6), alpha=1)
ax2 = sns.lineplot(x="Year", y="Rank", data=dt, legend=False,alpha=1)
sns.set(rc={"figure.dpi":100, 'savefig.dpi':300})
```
